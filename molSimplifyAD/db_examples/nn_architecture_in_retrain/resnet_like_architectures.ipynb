```python
import numpy as np
from molSimplifyAD.retrain.nets import ANN
```

```python
space = {'lr':0.01,
         'drop_rate': 0.1,
         'reg': 0.1,
         'batch_size': 16,
         'hidden_size': [200, 200],
         'beta_1': 0.95,
         'decay': 0.01,
         'res': False,
         'bypass': True,
         'amsgrad': False,
         'patience': 10,
         }
```

```python
model = ANN(space, input_len=100, lname=['homo', 'gap'], regression=True)
```
